Part 1: Biological Neural Response

```python
import numpy as np
import matplotlib.pyplot as plt

class BiologicalNeuron:
    """Simulates basic biological neuron characteristics"""
    def __init__(self):
        self.resting_potential = -70  # mV
        self.threshold = -55  # mV

    def simulate_response(self, input_current):
        """Simulates neuron response to input current"""
        # Initialize membrane potential to resting potential
        membrane_potential = self.resting_potential
        # Simulate the response over time
        time_steps = 100  # Adjust as needed
        membrane_potentials = []  # Initialize the list to store membrane potential values
        for t in range(time_steps):
            # Update membrane potential based on input current (simplified model)
            membrane_potential += input_current
            # Check if threshold is reached
            if membrane_potential >= self.threshold:
                # Generate an action potential (simplified representation)
                membrane_potential = 40  # Peak potential
            # Simulate decay of membrane potential (simplified model)
            membrane_potential -= 0.1  # Adjust decay rate as needed
            membrane_potentials.append(membrane_potential)  # Append value to list

        # Plot the membrane potential over time
        plt.plot(membrane_potentials)
        plt.xlabel("Time")
        plt.ylabel("Membrane Potential (mV)")
        plt.title("Biological Neuron Response")
        plt.show()

# Example usage
neuron = BiologicalNeuron()
neuron.simulate_response(1)  # Example input current
```

Part 2: Artificial Neuron Implementation

```python
import numpy as np

class ArtificialNeuron:
    """Implements artificial neuron for comparison"""
    def __init__(self, num_inputs):
        self.weights = np.random.randn(num_inputs)
        self.bias = 0

    def activate(self, inputs):
        """Implements artificial neuron activation"""
        # Calculate weighted sum of inputs
        weighted_sum = np.dot(self.weights, inputs) + self.bias
        # Apply activation function (example: sigmoid)
        output = 1 / (1 + np.exp(-weighted_sum))
        return output

# Example usage
neuron = ArtificialNeuron(3)  # Neuron with 3 inputs
inputs = np.array([0.5, 0.2, 0.8])  # Example input values
output = neuron.activate(inputs)
print(output)
```

```python
import numpy as np

def sigmoid(weighted_sum):
    return 1 / (1 + np.exp(-weighted_sum))

def relu(weighted_sum):
    return np.maximum(0, weighted_sum)  # Corrected for NumPy arrays

def tanh(weighted_sum):
    return np.tanh(weighted_sum)

# Example usage
weighted_sum = np.array([-2, -1, 0, 1, 2])  # Example input values

print("Sigmoid:", sigmoid(weighted_sum))
print("ReLU:", relu(weighted_sum))  # Now works correctly for arrays
print("Tanh:", tanh(weighted_sum))
```

Part 3: Information Integration Study

```python
import numpy as np
import matplotlib.pyplot as plt

class BiologicalNeuron:
    """Simulates basic biological neuron characteristics"""
    def __init__(self):
        self.resting_potential = -70  # mV
        self.threshold = -55  # mV

    def activate(self, input_current):
        """Simulates neuron response to input current"""
        membrane_potential = self.resting_potential
        time_steps = 100
        membrane_potentials = []
        for _ in range(time_steps):
            membrane_potential += input_current
            if membrane_potential >= self.threshold:
                membrane_potential = 40  # Action potential
            membrane_potential -= 0.1  # Decay
            membrane_potentials.append(membrane_potential)
        return membrane_potentials

class ArtificialNeuron:
    """Implements artificial neuron for comparison"""
    def __init__(self, num_inputs):
        self.weights = np.random.randn(num_inputs)
        self.bias = 0

    def activate(self, inputs):
        """Implements artificial neuron activation"""
        weighted_sum = np.dot(self.weights, inputs) + self.bias
        return 1 / (1 + np.exp(-weighted_sum))  # Sigmoid activation

# Constructing a simple feedforward network
class FeedforwardNetwork:
    """Simple feedforward network with biological neurons"""
    def __init__(self, num_neurons):
        self.neurons = [BiologicalNeuron() for _ in range(num_neurons)]

    def process(self, inputs):
        outputs = [neuron.activate(inp) for neuron, inp in zip(self.neurons, inputs)]
        return outputs

# Constructing a simple recurrent network
class RecurrentNetwork:
    """Simple recurrent network with biological neurons"""
    def __init__(self, num_neurons):
        self.neurons = [BiologicalNeuron() for _ in range(num_neurons)]
        self.previous_outputs = np.zeros(num_neurons)

    def process(self, inputs):
        outputs = []
        for i, neuron in enumerate(self.neurons):
            combined_input = inputs[i] + 0.5 * self.previous_outputs[i]  # Recurrent connection
            outputs.append(neuron.activate(combined_input))
        self.previous_outputs = np.array([out[-1] for out in outputs])  # Store last output for feedback
        return outputs

# Example input patterns
input_pattern1 = np.array([1, 2, 3])  # Different input values
input_pattern2 = np.array([3, 2, 1])

# Instantiate networks
ff_network = FeedforwardNetwork(3)
rec_network = RecurrentNetwork(3)

# Process inputs
ff_outputs = ff_network.process(input_pattern1)
rec_outputs = rec_network.process(input_pattern2)

# Plot results
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(ff_outputs[0], label='Neuron 1')
plt.plot(ff_outputs[1], label='Neuron 2')
plt.plot(ff_outputs[2], label='Neuron 3')
plt.title("Feedforward Network Response")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(rec_outputs[0], label='Neuron 1')
plt.plot(rec_outputs[1], label='Neuron 2')
plt.plot(rec_outputs[2], label='Neuron 3')
plt.title("Recurrent Network Response")
plt.legend()

plt.show()
```
